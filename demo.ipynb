```python
import polars as pl
from pathlib import Path

# Define your directory
parquet_dir = Path('/Volumes/One Touch/bigdata/data/gaia_partitioned')

# Define the expected columns
expected_cols = [
    'phot_g_mean_mag', 'phot_bp_mean_mag', 'phot_rp_mean_mag', 'bp_rp',
    'has_rvs', 'ra', 'dec', 'parallax', 'pmra', 'pmdec', 'source_id',
    'healpix_2'
]
expected_set = set(expected_cols)

# Iterate over each .parquet file
for file in parquet_dir.glob("*.parquet"):
    if file.name.startswith("._"):
        continue  # skip macOS metadata files

    try:
        df = pl.read_parquet(file, n_rows=1)  # only need schema, so read 1 row
        actual_set = set(df.columns)

        if actual_set != expected_set:
            missing = expected_set - actual_set
            extra = actual_set - expected_set

            print(f"\nFile: {file.name} does not match expected columns.")
            if missing:
                print(f"   Missing: {missing}")
            if extra:
                print(f"   Extra: {extra}")
        else:
            print(f" {file.name} - columns match.")

    except Exception as e:
        print(f"Error reading {file.name}: {e}")
```

```python
iau_traditional_names = [
    "Sirius", "Canopus", "Arcturus", "Vega", "Capella", "Rigel", "Procyon",
    "Achernar", "Betelgeuse", "Hadar", "Altair", "Acrux", "Aldebaran",
    "Spica", "Antares", "Pollux", "Fomalhaut", "Deneb", "Mimosa",
    "Regulus", "Adhara", "Shaula", "Castor", "Gacrux", "Bellatrix",
    "Elnath", "Miaplacidus", "Alnilam", "Alnair", "Alioth", "Dubhe",
    "Mirfak", "Wezen", "Sargas", "Kaus Australis", "Avior", "Alkaid",
    "Menkent", "Alphard", "Polaris", "Alphecca", "Gienah", "Alsephina"
    # ... you can add more if you want
]

import pandas as pd

# Assuming combined_with_names is your DataFrame

import re

def clean_name(name):
    if pd.isna(name):
        return ""
    # Remove common prefixes like '*', 'V*', 'V* ', etc., and strip spaces
    cleaned = re.sub(r'^[\*\sV\*]+', '', name).strip()
    return cleaned

def filter_iau_names(row):
    cleaned = clean_name(row['star_name'])
    for traditional_name in iau_traditional_names:
        if cleaned.lower() == traditional_name.lower():
            return True
    return False

iau_named_stars = stars[stars.apply(filter_iau_names, axis=1)]

print(iau_named_stars[['source_id', 'original_ext_source_id', 'star_name']])
```
